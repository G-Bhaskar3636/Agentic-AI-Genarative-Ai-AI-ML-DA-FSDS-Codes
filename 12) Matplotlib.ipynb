# Matplotlip from start

- There are two styles to make plots:
- 1) Pyplot style (simple way)
- 2) Object-Oriented (OO) style (professional way)
# Pyplot Method  Easy and Quick Way:
- Think of this like using your phone camera  just point and click!
- You use matplotlib.pyplot (we usually call it plt) to draw graphs directly

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python", "java", "c", "c++"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]
width = 0.2

p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("languange")
plt.ylabel("demand")
plt.title("languanr race")

plt.xticks(p+width/2, x, rotation = 45)

plt.bar(p, y, width, color = "r", label = "2020")
plt.bar(p1, z, width, color = "y", label = "2025")

plt.legend()
plt.show()
```

```python
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.ylabel("languange")
plt.xlabel("demand")
plt.title("languanr race")

plt.yticks(p+width/2, x, rotation = 45)

plt.barh(p, y, width, color = "r", label = "2020")
plt.barh(p1, z, width, color = "y", label = "2025")

plt.legend()
plt.show()
```

# Fill between plot

```python
x = np.linspace(0, 2*np.pi, 50)
s = np.sin(x)
c = np.cos(x)

s_upper = s + 0.05
s_lower = s - 0.05

c_upper = c + 0.05
c_lower = c - 0.05

plt.plot(x, s, color = "blue")
plt.fill_between(x, s_lower, s_upper, color = "blue", alpha = 0.2)

plt.plot(x, c, color = "orange")
plt.fill_between(x, c_lower, c_upper, color = "orange", alpha = 0.2)

plt.plot()
```
