- It always returns the user input as a string

```python
name = input("Enter your name: ")
print(f"Hello {name}")
```

```python
t, h, e = map(int, input("enter tel, hin, eng marks: ").split())
tot = t + h + e
avg = tot/3
print(f"Telugu = {t}, Hindi = {h}, English = {e}, Total = {tot}")
print(f"average = {avg}")
```

```python
c = float(input("temperature in Celsius: "))
f = (c*9/5)+32
print(f"temp in fahrenhit: {f}")
```

```python
sen = input("enter sentence: ").lower()
v, c, s = 0, 0, 0
vl, cl, = [], []
for i in sen:
    if i in ["a", "e", "i", "o", "u"]:
        v += 1
        vl.append(i)
    elif i == " ":
        s += 1
    else:
        c += 1
        cl.append(i)
print(f'''vowel count = {v}, {vl}
consonent count = {c}, {cl}
space count = {s}''')
```
