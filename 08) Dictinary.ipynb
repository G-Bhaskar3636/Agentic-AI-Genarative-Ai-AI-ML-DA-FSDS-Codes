```python
s = "banana"
d2 = {}
for i in s:
    if i in d2:
        d2[i] += 1
    else:
        d2[i] = 1
d2
```

```python
d1 = {'a': 10, 'b': 20, 'c': 5}
a = max(d1, key= d1.get)
a
```

```python
d2 = {'a': 10, 'b': 20, 'c': 5}
d = dict(sorted(d2.items(), key= lambda x:x[1]))
d
```

```python
a = {v:k for k, v in d.items()}
a
```

```python
d = ["apple", "banana", "cat", "dog", "elephant", "frog", "zebra"]
b = {}
for i in d:
    l = len(i)
    b.setdefault(l, []).append(i)
b
```

```python
l1 = ["a", "b", "c", "d"]
l2 = [1, 2, 3, 4]
d = {}
for k, v in zip(l1, l2):
    d[k] = v
d
```

```python
a = {"bhaskar":95, "anji": 96, "rakesh": 89, "sumanth": 85, "pradeep": 87}
b = dict(sorted(a.items(), key = lambda x:x[1], reverse = True)[:3])
b
```

```python
a = {"person1": {
        "name": "Alice",
        "details": {
            "age": 30,
            "city": "New York"
        }
    },
    "person2": {
        "name": "Bob",
        "details": {
            "age": 25,
            "city": "London"
        }
    }
}
a["person1"]["details"]["city"]
```

```python
b = { 'a': [1,2], 'b': [2,3] }
d = {}
dj = 97
for v in b.values():
    for i in v:
        if i not in d:
            d[i] = chr(dj)
            dj += 1
print(d)
```
