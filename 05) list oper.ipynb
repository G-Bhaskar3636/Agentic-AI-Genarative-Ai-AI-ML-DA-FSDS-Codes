```python
l = [1, 3, 4, 6, 8, 2, 4, 5, 2, 4, 6, 3, 6, 7, 9]
li = []
left, right = 0, len(l)-1
while left < right:
    if l[left] + l[right] == 10:
        li.append((l[left], l[right]))
        left += 1
        right -= 1
    elif l[left] + l[right] < 10:
        right -= 1
    else:
        left += 1
li
```

```python
l = [50, 60, 30, 50, 50, 70, 40]
d = dict()
for i in l:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
        h = 0
for k, v in d.items():
    if v > h:
        h = k
h
```
